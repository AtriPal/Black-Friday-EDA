```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing dataset

df_train = pd.read_csv('../input/black-friday/train.csv')
df_train.head()
```

# Problem Statement

A retail company 'ABC Private Limited' wants to understand the customer purchase behaviour (specifically purchase amount) against various products of different categories. They have shared purchase summary of variuos customers of selected high volume products from last month. The dataset also contains customer demographics (age, gender, maritial status, city type, stay in current city), product details (product_id and product category) and Total pruchase_amount from last month.

Now they want to build a model to predict the purchase amount of customer again products which will help them to create personalized offer for customers against different products.

```python
# import the test data 

df_test = pd.read_csv('../input/black-friday/test.csv')
df_test.head()
```

```python
# Merge both train and test data

df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
# Basic

df.info()
```

```python
df.describe()
```

```python
# drop user id

df.drop(['User_ID'],axis=1, inplace=True)
```

```python
df.head()
```

```python
# converting categorical features into numerical values
## like -- Gender, Age, city_category
### we will use MAP. we can map any data with any value

df['Gender'] = df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
## handle Age

df['Age'].unique() # unique give the unique values
```

```python
df['Age'] = df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6,'55+':7})
```

```python
df.head()
```

```python
## handle City_Category

df['City_Category'].unique()
```

```python
df['City_Category'] = df['City_Category'].map({'A':1, 'B':2, 'C':3})
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df.drop(['Stay_In_Current_City_Years'], axis=1, inplace=True)
df.head()
```

```python
## Missing value

df.isnull().sum()
```

```python
## Focus on replacing missing values
# (Product_cat 2, Product_cat 3)

df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
## Replace the missing value with mode

df['Product_Category_2'].mode()[0]
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
## Similarly for Product_Category_3

df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
df['Product_Category_3'].mode()[0]
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```
